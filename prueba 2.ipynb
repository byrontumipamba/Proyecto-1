```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *    
from sympy.plotting import plot_parametric
from mpl_toolkits.mplot3d import Axes3D
import math

fig = plt.figure()
ax = Axes3D(fig)

z = np.linspace(0, 10, 50)

def x(z):
  return (1/8)* z**4 + np.sin((3/4)*z**2) 
def y(z):
  return 2 * np.sin(1/4 * z**2)

ax.plot(x(z), y(z), z)
plt.show()
```
